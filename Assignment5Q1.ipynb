<b>Part B, C and D</b>


```python
label = pd.read_csv('/Users/beatrizruiz/Downloads/HW5/label.csv')
image_df = pd.DataFrame([])
for i in range(len(label.index)):
    img = label.iloc[i,:]
    images = mpimg.imread('/Users/beatrizruiz/Downloads/HW5/Digit/'+ str(img.values[0]) + '.jpg')
    pic = np.array(images)
    feature = pd.DataFrame(pic).stack().to_frame().T
    feature.columns = feature.columns.droplevel(0)
    image_df = image_df.append(feature, ignore_index=True)
    
image_df = label.join(image_df)
    
#image_df.head()



#print(label)
X = image_df[['name of the file']]
y = image_df['digit label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

my_RandomForest = RandomForestClassifier(n_estimators = 19, bootstrap = True, random_state=2)

my_RandomForest.fit(X_train,y_train)

y_predict = my_RandomForest.predict(X_test)

accuracy = accuracy_score(y_test,y_predict)

print('Accuracy Score for Random Forest:',accuracy)


```
