<b> Part A </b>

```python
#import linear regression, DecisionTreeClassifier, numpy, and pandas
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
```

```python
#read the data file and assign to pandas dataframe
heart_df = pd.read_csv('https://github.com/mpourhoma/CS4661/raw/master/Heart_s.csv')
```

<b> Part B</b>

```python
heart_df[0::10]
```

<b> Part C </b>

```python
#build feature matrix and label vector
heart_features = ['Age','RestBP','Chol','MaxHR','RestECG','Oldpeak']

X = heart_df[heart_features]
print(X.shape)
y = heart_df['AHD']
```

<b> Part D</b>

```python
#split dataset into training and testing sets 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X,y, test_size=0.25, random_state=6)
```

<b> Part E </b>

```python
#Knn classifier accuracy
k = 3
heart_knn = KNeighborsClassifier(n_neighbors=k)
heart_knn.fit(X_test,y_test)
print(X_train.shape)
print(y_train.shape)
heart_knn.fit(X_train,y_train)
y_predict = heart_knn.predict(X_test)
#print(y_predict)

#Calculate Accuracy Evaluation
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_predict)
print('KNN: ',accuracy)



# Logistic Regression Accuracy

heart_logreg = LogisticRegression()
heart_logreg.fit(X,y)


heart_logreg.fit(X_train, y_train)
y_predict_logreg = heart_logreg.predict(X_test)

accuracy_logreg = accuracy_score(y_test, y_predict_logreg)
print('Linear Regression',accuracy_logreg)

```

```python
#Decision Tree Classifier Accuracy with random_state = 5


heart_dectree = DecisionTreeClassifier(random_state=5)
#training a predictive model using the training dataset
heart_dectree.fit(X,y)

heart_dectree.fit(X_train,y_train)
y_predict_dt = heart_dectree.predict(X_test)

accuracy_dt = accuracy_score(y_test, y_predict_dt)
print('Decision Tree: ', accuracy_dt)


```

<p> The best classifier was the linear regression classifier with a 67% accuracy. The worst classifier was the decision tree classifier with 60% accuracy</p>
